```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=87adbeef-97af-430f-af9f-36c247f9f31d&as-searchtext=lapt'
page = requests.get(url)
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text,'html.parser')
```

```python
soup.title
```

```python
name = []
price = []
for a in soup.findAll('a',href=True, attrs={'class':'_31qSD5'}): #css class using inspect, using findAll method
  na = a.find('div', attrs={'class':'_3wU53n'})
  pr = a.find('div', attrs={'class':'_1vC4OE _2rQ-NK'})
  name.append(na.text)
  price.append(pr.text)
```

```python
#creating data frame 
import pandas as pd

df = pd.DataFrame(list(zip(name,price)))
df.columns = ['Name','Price']
```

```python
df
```

```python
df.to_csv('laptop_table.csv')
```
